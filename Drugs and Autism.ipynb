```python
import pandas as pd #used for working with dataframes—our dataset
import numpy as np #used to manipulate arrays
import matplotlib.pyplot as plt #for making visuals
import seaborn as sns #for making visuals too
```

```python
US18=pd.read_csv(r"C:\Users\melinadiaz\Documents\GitHub\us_18Q1.csv")
```

```python
drugUseCols=['CAN_REC_USE', 'CAN_MED_USE','CAN_USE', 'COKE_USE', 'CRACK_USE', 'MDMA_USE',
             'GHB_USE', 'INH_USE', 'SPEED_USE', 'METHAM_USE', 'NPFENT_USE', 'HEROIN_USE', 
             'KET_USE', 'MEPH_USE', 'LSD_USE', 'MUSH_USE', 'PCP_USE', 'MESC_USE', 'SPICE_USE',
             'SAL_USE', 'STER_USE', 'MENT_AUT']
```

```python
druguse=US18[drugUseCols].copy()
druguse
```

```python
len(drugUseCols)
```

```python
for i in range(21):
    druguse[drugUseCols[i]]= druguse[drugUseCols[i]].replace([1,2,3,4,5], [0,1,1,1,1])
druguse
```

```python
druguse[druguse['MENT_AUT']==1]
```

```python
drug_freq=druguse.groupby('MENT_AUT')[drugUseCols].mean()
drug_freq
```

```python
yes_aut = drug_freq.iloc[0].to_frame().reset_index()
yes_aut['MENT_AUT']=0
yes_aut.sort_values(0,ascending=False)
```

```python
no_aut = drug_freq.iloc[1].to_frame().reset_index().rename({1:0},
                                                    axis = 'columns')
no_aut['MENT_AUT']=1
no_aut
```

```python
drug_freq=no_aut.append(yes_aut)
drug_freq
```

```python
drug_freq1= drug_freq.drop(21, axis='index')
drug_freq1
```

```python
print(drug_freq.shape)
drug_freq1.shape
```

```python
g = sns.catplot(x = 'index',
                y = 0,
                hue = 'MENT_AUT',
                kind = 'bar',
                data = drug_freq1)
(g.set_axis_labels('Drug Type', 'Frequency Use')
  .set_xticklabels(['CannabisNMU', 'CannabisMU', 'Cannabis', 'Cocaine', 'Crack', 'MDMA', 'GHB', 'INH', 'Speed', 'Meth',
                    'Fentanyl', 'Heroin', 'Ketamine', 'MEPH', 'LSD', 'Shrooms','PCP', 'MESC', 'Spice', 'SAL', 'Steroids'], rotation=40)
)  
g.fig.suptitle("Autism and Drug Use")
```
